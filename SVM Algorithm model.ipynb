```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn import datasets 
iris=datasets.load_iris()
x= iris.data
y = iris.target
```

```python
names = iris.target_names
```

```python
df = pd.DataFrame(x, columns=iris.feature_names)
df['species']=iris.target #adding species as target to data frame
```

```python
df['species']=df['species'].replace(to_replace=[0,1,2], value=['setpsa', 'versicolor', 'virginica'])
df
```

```python
import seaborn as sns
sns.pairplot(data = df, hue='species') #visualize the data ,This creates a nice visualisation and helps us understand the data by summarising a large amount of data in a single figure
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
print(x_train.shape, y_train.shape)
```

```python
from sklearn.svm import SVC
mod = SVC(kernel='linear', random_state = 0)
```

```python
mod.fit(x_train,y_train)
```

```python
predictions = mod.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```
